### **Importando as bibliotecas**

```python
import pandas as pd
import numpy as np
```

### **Criando tabelas** 

```python
db = pd.DataFrame(index=['Segunda','Terça','Quarta','Quinta','Sexta','Sábado','Domingo'], columns=['Horas Trabalhadas','Bugs Corrigidos','Tarefas Concluídas'])
db.columns.name = 'Dias'

db.loc['Segunda'] = [6, 3, 5]
db.loc['Terça'] = [7, 2, 4]
db.loc['Quarta'] = [8, 1, 6]
db.loc['Quinta'] = [6, 4, 4]
db.loc['Sexta'] = [7, 3, 5]
db.loc['Sábado'] = [5, 2, 3]
db.loc['Domingo'] = [4, 1, 2]

db
```

```python
db = pd.DataFrame(columns=['Dias','Horas Trabalhadas','Bugs Corrigidos','Tarefas Concluídas'])

db.loc[0] = ['Segunda', 6, 3, 5]
db.loc[1] = ['Terça', 7, 2, 4]
db.loc[2] = ['Quarta', 8, 1, 6]
db.loc[3] = ['Quinta', 6, 4, 4]
db.loc[4] = ['Sexta', 7, 3, 5]
db.loc[5] = ['Sábado', 5, 2, 3]
db.loc[6] = ['Domingo', 4, 1, 2]
db

```

##### **Adcionando Total de Horas**

```python
db.loc[7] = db[['Horas Trabalhadas','Bugs Corrigidos','Tarefas Concluídas']].sum(axis=0)
db.at[7, 'Dias'] = '*Totais*'

db
```

##### **Adcionando Medias**

```python
db.loc[8] = db[['Horas Trabalhadas','Bugs Corrigidos','Tarefas Concluídas']].mean(axis=0)
db.at[8, 'Dias'] = '*Medias*'

db
```

##### **Adcionando Produtividade Diária**

```python
db['Produtividade Diária '] = db['Tarefas Concluídas'] / db['Horas Trabalhadas']
db
```
