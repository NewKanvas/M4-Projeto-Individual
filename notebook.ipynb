### **Importando as bibliotecas**

```python
import pandas as pd
import numpy as np
```

### **Criando tabelas** 

```python
db = pd.DataFrame(columns=['Dias','Horas Trabalhadas','Bugs Corrigidos','Tarefas Concluídas'])

db.loc[0] = ['Segunda', 6, 3, 5]
db.loc[1] = ['Terça', 7, 2, 4]
db.loc[2] = ['Quarta', 8, 1, 6]
db.loc[3] = ['Quinta', 6, 4, 4]
db.loc[4] = ['Sexta', 7, 3, 5]
db.loc[5] = ['Sábado', 5, 2, 3]
db.loc[6] = ['Domingo', 4, 1, 2]
db

```

##### **Adcionando Total de Horas**

```python
db.loc[7] = db[['Horas Trabalhadas','Bugs Corrigidos','Tarefas Concluídas']].sum()
db.at[7, 'Dias'] = 'Totais'

db
```

##### **Adcionando Médias**

```python
##### **Adcionando Médias**
db.loc[8] = db.loc[0:5, ['Horas Trabalhadas', 'Bugs Corrigidos', 'Tarefas Concluídas']].mean()
db.at[8, 'Dias'] = 'Medias Totais'

db
```

##### **Adcionando Produtividade Diária**

```python
db['Produtividade Diária'] = db['Tarefas Concluídas'] / db['Horas Trabalhadas']
db
```

```python
db2 = db.set_index(['Dias'])
db2.drop(['Totais'],inplace= True)
db2
```

### **Plotando**

#### Plotando o Dataframe

```python
db2.plot.barh()
```

#### Plotando Horas Trabalhadas

```python
db2[['Horas Trabalhadas']].plot.bar(rot=35).axhline(y=db2.loc['Medias Totais', 'Horas Trabalhadas'], color='orange', linestyle='--',)
```

#### Plotando Bugs Corrigidos

```python
db2[['Bugs Corrigidos']].plot.bar(rot=35).axhline(y=db2.loc['Medias Totais', 'Bugs Corrigidos'], color='orange', linestyle='--')
```

#### Plotando Tarefas Concluídas

```python
db2[['Tarefas Concluídas']].plot.bar(rot=35).axhline(y=db2.loc['Medias Totais', 'Tarefas Concluídas'], color='orange', linestyle='--')
```

#### Plotando Produtividade Diária

```python
db2[['Produtividade Diária']].plot(rot=35, marker='o', grid=True).axhline(y=db2.loc['Medias Totais', 'Produtividade Diária'], color='orange', linestyle='--')
```
